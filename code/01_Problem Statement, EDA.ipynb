## Image Classification

* First, we will carry out image classification to predict the class of an object in an image, i.e. which of the 26 alphabets is a hand signing. The input will be an image with a single object (a photograph), while output will be the class label (which alphabet).
* We will mainly be exploring Transfer Learning with Pre-trained Models, the models to be explored include:

| Model          | Characteristics of Model                             |
|-------------------|-----------------------------------------------------|
| VGG16            | • Deep convolutional neural network (CNN) model supporting 16 layers.<br>• Utilizes smaller 3x3 convolutional filters, reduces tendency for overfitting |
|InceptionV3         |• Uses multiple filters of different sizes in parallel to capture features at different scales<br>• Has low computational demands compared to other architectures |                                                    |
| MobileNetV2      | • Uses Depthwise Separable Convolution which reduces complexity of model.<br>• Adopts an inverted residual structure<br>• Well balanced between model size, computational efficiency, and accuracy |
| EfficientNetB0    | • Achieves state-of-the-art accuracy with fewer parameters (computationally efficient)<br>• Provides a good balance between model size and performance.

<br>

* For our image classification model, the industry standard of **70% accuracy rate** can be used for our model to be deemed successful.  Sign language is a form of communication where the recipient often plays an active role in interpreting the message. In this case, a high accuracy is not imperative as users have the flexibility to clarify or seek additional context in case of misinterpretation. Additionally, the interpretation of hand gestures can vary significantly based on factors like lighting conditions, hand orientation, and background clutter. Considering these factors, a 70% accuracy rate may represent a reasonable level of performance.<br><br>
* For model comparison, accuracy and loss metrics serve as important indicators of the model's performance. While accuracy indicates the overall correctness of predictions, loss helps in fine-tuning the model during training to improve its accuracy.
    * **Accuracy**: Accuracy is chosen because it is a straightforward metric that measures the percentage of correctly classified instances. Also, the data for our alphabet classes are very balanced. Accuracy can provide a clear understanding of how well the model is performing overall in terms of correctly recognizing the sign language gestures. 
    * **Loss**: Loss functions, such as cross-entropy loss, provide insight into how well the model's predictions align with the ground truth alphabet labels. Minimizing the loss helps improve the model's ability to make accurate predictions.

## Object Detection
* Object detection architectures are mainly categorised into 2 groups:
    * Two-Stage Learning: [Involves extracting the Region of Interest (ROI) and then classifying and regressing the ROIs](https://www.v7labs.com/blog/object-detection-guide). Some examples include: R-CNN, Fast-RCNN, Faster-RCNN, etc.
    * Single-Shot Learning: Able to directly predict object bounding boxes and class labels in a single pass through the network. Some examples include: YOLO (You Only Look Once), SSD (Single Shot MultiBox Detector).
<br> 
<br>
* While two-stage object detection models are generally more accurate, they require way [more computational power and are slower](https://towardsdatascience.com/the-basics-of-object-detection-yolo-ssd-r-cnn-6def60f51c0b). Hence, for this project, for the purpose of enabling end users to do image and video-based detection through a mobile phone, we will try out the single-shot detection models instead.

| Model          | Characteristics of Model                             |
|-------------------|----------------------------------------------------|
| YOLOv8            | • State-of-the-art (SOTA) model that is [fast, accurate and easy to use](https://github.com/ultralytics/ultralytics)<br>• Uses anchor-free detection and new convolutional layers to make predictions more accurate<br>• [Greatly improved MAP value for slightly more training time (compared to YOLOv5)](https://medium.com/mlearning-ai/yolo-v8-the-real-state-of-the-art-eda6c86a1b90)|


<br>

* For our object detection model, the metrics used include:
    * **Mean Average Precision (mAP)**: Summarises the precision and recall of the detected objects across different confidence thresholds and provides a single value that quantifies overall detection performance
    * **Precision and Recall**: Precision measures the accuracy of object detection by calculating the ratio of true positives to the total number of predicted positives (true positives plus false positives). Recall measures the model's ability to detect all instances of a given object class by calculating the ratio of true positives to the total number of ground truth positives (true positives plus false negatives).

### Loading Train and Test Data

The below constants / parameters were decided following earlier model experiments:

```python
BATCH_SIZE = 64
IMG_SIZE = 224
EPOCHS = 50
DROPOUT_RATE = 0.2
```

Let's load the train and test data using the flow_from_directory input pipeline.<br>
* After trial-and-error, I decided to omit using the `preprocessing_function = preprocess_input` and replaced it with `rescale = 1.0/225.0` instead. This led to a vast improvement in results - testing accuracy remained similar, but training accuracy improved by almost 20%. Also, as preprocess_input scales to [-1, 1], it was less favourable in terms of interpretability + plotting of images, brightness levels, etc.

* I decided to use class_mode as `'sparse'` instead of 'categorical', because of the following advantages:
    * Simplicity: Sparse labels are straightforward and easy to understand. Each image is associated with a single integer, making it clear which class it belongs to.
    * Memory Efficiency: Sparse labels require less memory compared to one-hot encoded labels because you only need to store a single integer per image, regardless of the number of classes.

```python
# Define the rescaling factor (typically 1.0/255 for RGB images)
rescale_factor = 1.0 / 255.0

# Load the Train and Validation data
train_data = ImageDataGenerator(rescale=rescale_factor).flow_from_directory('../data/asl_alphabet/train/', 
                                                                                             target_size=(IMG_SIZE,IMG_SIZE), 
                                                                                             class_mode='sparse',
                                                                                             seed=42
                                                                                            )

test_data = ImageDataGenerator(rescale=rescale_factor).flow_from_directory('../data/asl_alphabet/test/', 
                                                                           target_size=(IMG_SIZE,IMG_SIZE),
                                                                           class_mode='sparse',
                                                                           seed=42
                                                                                          )
```

## 3.2 Preview of Data Images
Let's plot 9 images selected randomly from the 'C' class folder for train dataset.

```python
def display_img_grid(data, category, number, grid_cols=3, random_state=None):
    class_indices = data.class_indices
    class_index = class_indices[category]
    displayed_images = 0

    np.random.seed(random_state)  # Set the random seed

    # Collect all class images
    class_images = []

    for images, labels in data:
        class_images.extend(images[labels == class_index])
        if len(class_images) >= number:
            break

    np.random.shuffle(class_images)
    
    fig, axs = plt.subplots(number // grid_cols, grid_cols, figsize=(10, 10))
    fig.tight_layout()

    for i, image in enumerate(class_images[:number]):
        row = i // grid_cols
        col = i % grid_cols
        axs[row, col].imshow(image)
        axs[row, col].axis('off')

    plt.show()
```

```python
display_img_grid(train_data, 'C', 9, grid_cols=3, random_state=45)  
```

## 3.3 Investigate Pixel Intensity (Brightness Range)
As we have several images with darker lighting, let's calculate and plot Pixel Intensity (Brightness Range) for the training dataset:

```python
# Initialize empty lists to store brightness values for train and test data
train_brightness_values = []
test_brightness_values = []

# Iterate through the batches of train_data
for batch_images, _ in train_data:
    # Calculate the average brightness for each image in the batch
    batch_brightness = np.mean(np.mean(batch_images, axis=(1, 2)), axis=1)
    # Append batch_brightness to the train_brightness_values list
    train_brightness_values.extend(batch_brightness)

    # Break the loop when all batches have been processed
    if len(train_brightness_values) >= len(train_data) * train_data.batch_size:
        break

# Iterate through the batches of test_data
for batch_images, _ in test_data:
    # Calculate the average brightness for each image in the batch
    batch_brightness = np.mean(np.mean(batch_images, axis=(1, 2)), axis=1)
    # Append batch_brightness to the test_brightness_values list
    test_brightness_values.extend(batch_brightness)

    # Break the loop when all batches have been processed
    if len(test_brightness_values) >= len(test_data) * test_data.batch_size:
        break
```

```python
# Calculate and print the brightness range
brightness_min = np.min(train_brightness_values)
brightness_max = np.max(test_brightness_values)
print(f"Brightness Range: {brightness_min} - {brightness_max}")
```

```python
# Plot the brightness histograms for both train and test data on the same plot
plt.figure(figsize=(8, 6))
plt.hist(train_brightness_values, bins=50, color='blue', alpha=0.5, label='Train Data')
plt.hist(test_brightness_values, bins=50, color='red', alpha=0.5, label='Test Data')
plt.title("Brightness Distribution of Training and Test Set Images")
plt.xlabel("Brightness Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

### Findings
* From the plot above, we can see that train and test brightness values are normally distributed, with a left-skewed distribution.
* The distribution is left-skewed because is has a long tail on the left side of the peak, meaning that at higher brightness values, there is like to be more number of images where their brightness levels observed is higher than average.
* As a large proportion of the images already have quite a low brightness range, we note this and make sure not to further darken the images during data augmentation.
* We also take note of the brightness range of 0.09-0.81, this can be reviewed again post-augmentation to see the variation.
